```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
import warnings
warnings.simplefilter(action='ignore',category=FutureWarning)
```

```python
df =pd.read_csv('iris.csv')
```

```python
df.sample(5)
```

```python
df.info()
```

```python
df['species']=df['species'].str.replace('Iris-','')
```

```python
df.describe()
```

```python
sns.heatmap(df.corr(),annot =True)
```

```python
df['sepal_length'].plot(kind='line',legend=True,marker ='o',color
                       ='r',figsize =(10,5),grid =True,rot=45)
df['sepal_width'].plot(kind='line',legend=True,marker ='o',color
                       ='b',figsize =(10,5),grid =True,rot=45)
plt.title('Sepal Length & Width',pad =10,loc ='center',fontdict={'fontsize':20,'color':'r','verticalalignment':'baseline'})
plt.xlabel('Count',labelpad=20,fontdict={'fontsize':15,'color':'r','verticalalignment':'baseline'})
plt.ylabel('Length/Width',labelpad=20,fontdict={'fontsize':15,'color':'b','verticalalignment':'baseline'})
```

## Sepal length is greater than Sepal width

```python
df['petal_length'].plot(kind='line',legend=True,marker ='o',color
                       ='r',figsize =(10,5),grid =True,rot=45)
df['petal_width'].plot(kind='line',legend=True,marker ='o',color
                       ='b',figsize =(10,5),grid =True,rot=45)
plt.title('Petal Length & Width',pad =10,loc ='center',fontdict={'fontsize':20,'color':'r','verticalalignment':'baseline'})
plt.xlabel('Count',labelpad=20,fontdict={'fontsize':15,'color':'r','verticalalignment':'baseline'})
plt.ylabel('Length/Width',labelpad=20,fontdict={'fontsize':15,'color':'b','verticalalignment':'baseline'})
```

## Petal length is greater than sepal width

```python
df['petal_length'].plot(kind='line',legend =True,
                       marker ='o',color ='g',figsize=(10,5),grid=True,rot =45)
df['petal_width'].plot(kind='line',legend =True,
                       marker ='.',color ='y',figsize=(10,5),grid=True,rot =45)
df['sepal_length'].plot(kind='line',legend =True,
                       marker ='*',color ='r',figsize=(10,5),grid=True,rot =45)
df['sepal_width'].plot(kind='line',legend =True,
                       marker ='p',color ='b',figsize=(10,5),grid=True,rot =45)
plt.title('Petal/ Petal Length & Width',pad =10,loc ='center',fontdict={
    'fontsize':20,'color':'r','verticalalignment':'baseline'})
plt.xlabel('Count',labelpad =10,fontdict={
    'fontsize':20,'color':'r','verticalalignment':'baseline'})
plt.ylabel('Length/width',labelpad =10,fontdict={
    'fontsize':20,'color':'r','verticalalignment':'baseline'})
```
